# Dataloader Walkthrough

Build a custom DataLoader class in Python to understand how real data pipelines start from basic file handling and structured method design.
Instead of using external libraries, we are going implement everything with core Python so the logic stays clear and easy to debug.

## Learning Goals
* define a reusable DataLoader class
* load text or CSV data from a file
* validate that data exists before processing it
* preview a few rows with sample()
* split the dataset into train and test parts
* calculate simple dataset stats without pandas

# Code

```python
class DataLoader:

    def __init__(self, path):
        self.path = path 
        self.data = None

    def load(self):
        with open(self.path, "r") as file:  
            content = file.read()
            self.data = content 
            return self
            
    def validate(self):
        if not self.data: 
            raise ValueError("Data is empty or not loaded")
        return self               
    
    def sample(self, n):
        if self.data is None:   
            raise ValueError("Please load the data")
        lines = self.data.splitlines()
        return lines[:n]
    
    def split(self, ratio):
        if self.data is None:
            raise ValueError("Please load the data")
        lines = self.data.splitlines() 
        split = int(len(lines)*ratio)     ## 

        train_data = lines[:split]
        test_data = lines[split:]

        return train_data, test_data
    
    def stats(self):
        if self.data is None:
            raise ValueError("Please load the data")                    
        lines = self.data.splitlines()
        find_column = lines[0].split(",")
        column = len(find_column)

        rows = len(lines)
        return {"rows": rows,
                "columns" : column
                }

loader1 = DataLoader("titanic.csv")
loader1.load()
print(loader1.stats())
```

## Lets understand Step by Step

## Step 1 - Define the class dataloader path

```python
class DataLoader:

    def __init__(self, path):
        self.path = path 
        self.data = None
```

## Step 2 - Method Load(): Reads the CSV/text file and stores the content in memory.

```python
def load(self):
        with open(self.path, "r") as file:  
            content = file.read()
            self.data = content 
            return self
```

## Step 3 - Method validate(): Checks whether data is loaded; raises a clear error for invalid cases.

```python
def validate(self):
        if not self.data:  # check if self.data is empty return False
            raise ValueError("Data is empty or not loaded")
        return self
```

## Step 4 - Method sample(n): Returns the first n lines for quick inspection

```python
def sample(self, n):
        if self.data is None:   # check is data is None raise error
            raise ValueError("Please load the data")
        lines = self.data.splitlines()
        return lines[:n]
```

## Step 5 - Method split(ratio): Divides data into two parts based on the given ratio (train/test style).

```python
def split(self, ratio):
        if self.data is None:
            raise ValueError("Please load the data")
            
        '''get split lines then find 
          the index of split for train_dat and test_data
        '''
        lines = self.data.splitlines() 
        split = int(len(lines)*ratio)     ## 

        train_data = lines[:split]
        test_data = lines[split:]

        return train_data, test_data
```

## Step 6- Method stats(): Returns total rows and number of columns.

```python
def stats(self):
        if self.data is None:
            raise ValueError("Please load the data")                    
        """ finding row and columns """
        lines = self.data.splitlines()
        find_column = lines[0].split(",")
        column = len(find_column)

        rows = len(lines)
        return {"rows": rows,
                "columns" : column
                }
```
