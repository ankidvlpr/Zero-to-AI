```python
from pathlib import Path
```

## 1. Define The Class

In a notebook, the full class should stay in one code cell.
That makes the notebook easy to run from top to bottom with `Run All`.

```python
class DataLoader:
    def __init__(self, path):
        self.path = Path(path)
        self.data = None

    def load(self):
        with self.path.open("r", encoding="utf-8") as file:
            self.data = file.read()
        return self

    def validate(self):
        if self.data is None or not self.data.strip():
            raise ValueError("Data is empty or not loaded")
        return self

    def sample(self, n):
        if self.data is None:
            raise ValueError("Please load the data")
        lines = self.data.splitlines()
        return lines[:n]

    def split(self, ratio):
        if self.data is None:
            raise ValueError("Please load the data")
        if not 0 < ratio < 1:
            raise ValueError("Ratio must be between 0 and 1")

        lines = self.data.splitlines()
        split_index = int(len(lines) * ratio)
        train_data = lines[:split_index]
        test_data = lines[split_index:]
        return train_data, test_data

    def stats(self):
        if self.data is None:
            raise ValueError("Please load the data")

        lines = self.data.splitlines()
        columns = len(lines[0].split(","))
        rows = len(lines)
        return {"rows": rows, "columns": columns}
```

## 2. Choose The Dataset Path

The dataset file is stored in the same folder as this notebook, so the relative path stays simple.

```python
possible_paths = [
    Path("titanic.csv"),
    Path("week07-Jupyter_DataTools/notebooks/dataloader/titanic.csv"),
]

for candidate in possible_paths:
    if candidate.exists():
        DATA_PATH = candidate
        break
else:
    raise FileNotFoundError("Could not find titanic.csv")

DATA_PATH
```

## 3. Load And Validate The Data

This step reads the file content into memory and checks that it is not empty.

```python
loader = DataLoader(DATA_PATH).load().validate()
type(loader.data), len(loader.data.splitlines())
```

## 4. Preview A Small Sample

The `sample()` method is useful when you want to quickly inspect the first few rows.

```python
loader.sample(3)
```

## 5. Split The Dataset

Here we divide the loaded data into two parts using an 80/20 split.

```python
train_data, test_data = loader.split(0.8)
len(train_data), len(test_data)
```

## 6. Check Basic Stats

This method counts rows and columns using core Python only.

```python
loader.stats()
```
