### Multiplying a (32x10) matrix by a (10x5) weight matrix

```python
import numpy as np
from numpy.linalg import *
```

#### Generating (32x10) matrix and (10x5) by numpy random

```python
rng = np.random.default_rng()
```

```python
array1 = rng.integers(1,10, size=(32,10))
```

```python
array2 = rng.integers(1,10,size=(10,5))
```

```python
print(array1)
```

```python
print(array2)
```

### Using dot and @ and matmul

```python
print(np.dot(array1,array2))  #dot product
```

```python
print(array1@array2) # using @ for multipication of matrix
```

```python
print(np.matmul(array1,array2))   #matmul matrix multipication
```

**(@,dot,matmul) uses for multiplication of matrix and vectors**

## Shape of array/matrix

```python
print("shape of array1:",array1.shape)  #array1 = (32x10),
print("shape of array2:",array2.shape)  #array2 = (10x5) 
```

```python
result = array1 @ array2  # (32,10) @ (10,5) -> inner 10 matches, so output shape becomes (32,5)
print("shape of result:", result.shape)
```

**rule** - matrix multiplication is valid when the columns of the first matrix match the rows of the second matrix.

## Covariance matrix


```python
hours = [2, 4, 6, 8, 10]
scores = [50, 60, 75, 85, 95]
```

```python
cov_matrix = np.cov(hours,scores)
```

```python
print("Covariance Matrix:\n",cov_matrix)
```

## Norm types

- `ord=1` -> L1 norm
- `ord=2` -> L2 norm
- `ord=np.inf` -> infinity norm
- `ord='fro'` -> Frobenius norm for matrices


```python
vector = np.array([3,4])
matrix = np.array([[1,2,3],
                   [2,3,4]])

print("Default norm:", np.linalg.norm(vector))
print("L1 norm:", np.linalg.norm(vector, ord=1))
print("L2 norm:", np.linalg.norm(vector, ord=2))
print("Infinity norm:", np.linalg.norm(vector, ord=np.inf))
print("Frobenius norm:", np.linalg.norm(matrix, ord='fro'))
```

## Mini summary

- Matrix multiplication: `(32,10) @ (10,5) -> (32,5)`
- `@`, `np.dot()`, and `np.matmul()` can all be used for matrix multiplication
- Covariance matrix shows how two variables vary together
- Norms measure vector or matrix size in different ways
