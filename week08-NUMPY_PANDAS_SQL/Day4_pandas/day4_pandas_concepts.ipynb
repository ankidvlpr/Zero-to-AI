# Day 4 Pandas Concepts

Simple notes with a few clean examples. This notebook focuses on understanding the idea first.

```python
import pandas as pd
from pathlib import Path

base_path = Path(".")
csv_path = base_path / "pokemonDB_dataset.csv"
json_path = base_path / "pokemonDB_dataset.json"
```

## 1. Series from a list

A `Series` is a one-dimensional labeled array. You can think of it like one column with an index.

```python
data = [23, 13, 223, 232, 334, 23]
series = pd.Series(data, index=["a", "b", "c", "d", "e", "f"])

print(series)
print()
print(series[series > 200])
```

## 2. Series from a dictionary

When we use a dictionary, keys become the index labels.

```python
portfolio = {"Day1": 100, "Day2": 300, "Day3": 200}
portfolio_series = pd.Series(portfolio)

print(portfolio_series)
print()
print("loc Day1 ->", portfolio_series.loc["Day1"])
print("iloc 1 ->", portfolio_series.iloc[1])
```

## 3. DataFrame basics

A `DataFrame` is a 2D table with rows and columns.

```python
students = pd.DataFrame({
    "Name": ["Person1", "Person2", "Person3"],
    "Age": [23, 24, 25],
})

students["Job"] = ["Cook", "Cashier", "Driver"]
new_row = pd.DataFrame([{"Name": "Person4", "Age": 24, "Job": "Scientist"}])
students = pd.concat([students, new_row], ignore_index=True)

students
```

## 4. Read CSV and JSON

CSV usually loads directly. This JSON file is nested, so `.T` makes it easier to read as rows.

```python
pokemon_df = pd.read_csv(csv_path)
pokemon_json_df = pd.read_json(json_path).T.reset_index().rename(columns={"index": "Pokemon"})

print(pokemon_df.shape)
print(pokemon_json_df.shape)
pokemon_df.head(2)
```

## 7. `loc` vs `iloc`

- `loc` uses labels
- `iloc` uses positions

```python
student_marks = pd.DataFrame({
    "Name": ["Aman", "Riya", "Kunal", "Sara"],
    "Marks": [88, 92, 79, 95],
    "City": ["Delhi", "Pune", "Noida", "Jaipur"]
}, index=["s1", "s2", "s3", "s4"])

# loc example 1
student_marks.loc["s2"]
```

```python
# loc example 2
student_marks.loc[["s1", "s3"], ["Name", "Marks"]]
```

```python
# iloc example 1
student_marks.iloc[1]
```

```python
# iloc example 2
student_marks.iloc[0:2, 0:2]
```

## 8. Filtering rows

We can convert height text to a number and then filter the rows we want.

```python
pokemon_df["Height_m"] = pokemon_df["Height"].str.split().str[0].astype(float)
filtered_df = pokemon_df[(pokemon_df["Type"] == "Grass") & (pokemon_df["Height_m"] >= 2)]

filtered_df[["Pokemon", "Type", "Height", "Height_m"]].head()
```

## 9. Aggregate functions

Aggregate functions give one summary value from many values.

```python
pokemon_df[["HP Base", "Attack Base", "Speed Base"]].mean()
```

```python
pokemon_df["Height_m"].describe()
```

## 10. `groupby()`

`groupby()` lets us summarize values for each category.

```python
pokemon_df["Catch Rate_num"] = pokemon_df["Catch Rate"].str.extract(r"(\d+)").astype("Int64")
pokemon_df.groupby("Type")["Catch Rate_num"].mean().sort_values(ascending=False).head(10)
```

## 11. Data cleaning basics

These are common cleaning steps: remove missing rows, fill missing values, replace wrong text, and remove duplicates.

```python
dirty_df = pd.DataFrame({
    "Name": ["A", "B", "B", None],
    "Type": ["Fire", "Dark", "Dark", None],
    "Score": [10, None, None, 20]
})

print("Original")
print(dirty_df)
print()

print("dropna on Type")
print(dirty_df.dropna(subset=["Type"]))
print()

print("fillna")
print(dirty_df.fillna({"Type": "Unknown", "Score": 0}))
print()

print("replace")
print(dirty_df.replace({"Dark": "DARK"}))
print()

print("drop_duplicates")
print(dirty_df.drop_duplicates())
```
