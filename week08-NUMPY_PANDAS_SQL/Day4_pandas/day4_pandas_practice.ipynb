This notebook keeps the same topics but gives more hands-on examples and ready solutions.

```python
import pandas as pd
from pathlib import Path

base_path = Path(".")
csv_path = base_path / "pokemonDB_dataset.csv"
json_path = base_path / "pokemonDB_dataset.json"
pokemon_df = pd.read_csv(csv_path)
pokemon_indexed = pd.read_csv(csv_path, index_col="Pokemon")
```

## Practice 1 - Series from a list

Task: Create a series with custom index labels and print only values greater than 200.

```python
# Solution
values = [23, 13, 223, 232, 334, 23]
series = pd.Series(values, index=["a", "b", "c", "d", "e", "f"])

print(series)
print()
print(series[series > 200])
```

## Practice 2 - Series from a dictionary

Task: Build a series from a dictionary, increase Day1 by 50, then access one value with `loc` and one with `iloc`.

```python
# Solution
portfolio = {"Day1": 100, "Day2": 300, "Day3": 200}
portfolio_series = pd.Series(portfolio)
portfolio_series.loc["Day1"] += 50

print(portfolio_series)
print("loc Day1 ->", portfolio_series.loc["Day1"])
print("iloc 2 ->", portfolio_series.iloc[2])
```

## Practice 10 - Filtering rows

Task: Find Pokemon with type Grass and height greater than or equal to 2 meters.

```python
# Solution
pokemon_df["Height_m"] = pokemon_df["Height"].str.split().str[0].astype(float)
filtered_df = pokemon_df[(pokemon_df["Type"] == "Grass") & (pokemon_df["Height_m"] >= 2)]

print(filtered_df[["Pokemon", "Type", "Height", "Height_m"]].head(10))
```

## Practice 11 - Aggregate functions

Task: Count values and find mean, sum, min, and max for height.

```python
# Solution
height_only = pokemon_df.copy()
height_only["Height"] = height_only["Height"].str.split().str[0].astype(float)

print(height_only.count())
print()
print("mean ->", height_only["Height"].mean())
print("sum ->", height_only["Height"].sum())
print("min ->", height_only["Height"].min())
print("max ->", height_only["Height"].max())
```

## Practice 12 - `groupby()`

Task: Convert catch rate to a number and summarize it by Pokemon type.

```python
# Solution
group_df = pokemon_df.copy()
group_df["Catch Rate"] = group_df["Catch Rate"].str.extract(r"(\d+)").astype("Int64")
group = group_df.groupby("Type")["Catch Rate"]

print(group.mean())
print()
print(group.sum())
print()
print(group.min())
```

## Practice 13 - Data cleaning

Task: Try common cleaning operations on a small messy DataFrame.

```python
# Solution
clean_df = pd.DataFrame({
    "Pokemon": ["A", "B", "B", None],
    "Type": ["Dark", "Dark", "Dark", None],
    "Score": [10, None, None, 20]
})

print("Original")
print(clean_df)
print()

print("dropna(subset=[\"Type\"])")
print(clean_df.dropna(subset=["Type"]))
print()

print("fillna")
print(clean_df.fillna({"Type": "Unknown", "Score": 0}))
print()

print("replace")
print(clean_df.replace({"Dark": "DARK"}))
print()

print("drop_duplicates")
print(clean_df.drop_duplicates())
```
