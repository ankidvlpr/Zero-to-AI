```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head(5)
```


#### Survival rate by Sex

```python
df.groupby("Sex")["Survived"].mean()
```

#### Average Age by Pclass

```python
df.groupby("Pclass")["Age"].mean()
```

#### Count by Embarked

```python
df.groupby("Embarked")["PassengerId"].count()
```

### Joining Two dataframe (merge & Join)

```python
df1 = pd.read_csv("LOTR.csv")
df2 = pd.read_csv("LOTR 2.csv")
```

```python
df1.merge(df2, how = "inner")
```

```python
df1.merge(df2, how ="left")
```

```python
df1.set_index("FellowshipID").join(df2.set_index("FellowshipID"), lsuffix = "_left", rsuffix = "_right", how = "outer")
```

### aggregations

```python
df.groupby("Sex").agg({
    "Survived": "mean",
    "Pclass": "mean",
    "Embarked": "count"
})
```

```python
df.groupby("Embarked").agg({
    "Age": "mean",
    "Fare": "mean"
})
```
