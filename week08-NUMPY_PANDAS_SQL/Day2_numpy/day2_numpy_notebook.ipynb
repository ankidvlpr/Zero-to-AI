Practice notebook for Day 2 NumPy topics: loop vs vectorized timing, broadcasting, dot product, and Titanic data analysis with NumPy.


```python
import csv
from pathlib import Path

import numpy as np
```

## Part 1: Loop vs Vectorized Speed Compare

Compare the speed of a Python loop and a NumPy vectorized operation for the same task.


```python
arr = np.arange(1_000_000)

%timeit result = [x * 2 for x in arr]
%timeit arr * 2
```

## Part 2: Broadcasting Example 1

Broadcasting with shape `(3,)` and shape `(3, 1)`.


```python
array1 = np.array([1, 2, 3])
array2 = np.array([[10], [20], [30]])

print(array1, array1.shape)
print(array2, array2.shape)
print(array1 + array2)
```

## Part 3: Broadcasting Example 2

Broadcasting with shape `(3, 1)` and shape `(4,)`.


```python
array1 = np.array([[1], [2], [3]])
array2 = np.array([100, 200, 300, 400])

print(array1, array1.shape)
print(array2, array2.shape)
print(array1 + array2)
```

## Part 4: Broadcasting Example 3

Broadcasting a 2D array with a 1D scaling array.


```python
image = np.array([[50, 100, 150], [60, 110, 160]])
scale = np.array([1.0, 0.5, 0.2])

print(image, image.shape)
print(scale, scale.shape)
print(image * scale)
```

## Part 5: Dot Product

First compute the dot product manually, then compare it with `np.dot()`.


```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

manual_dot = np.sum(array1 * array2)
numpy_dot = np.dot(array1, array2)

print(manual_dot)
print(numpy_dot)
```

## Part 6: Titanic Data Load

Load the Titanic CSV file and convert selected columns into NumPy arrays.


```python
possible_paths = [
    Path('..') / '..' / 'week05-Dataloader_class' / 'titanic.csv',
    Path('week05-Dataloader_class') / 'titanic.csv',
]

for path in possible_paths:
    if path.exists():
        csv_path = path
        break

rows = list(csv.DictReader(open(csv_path, encoding='utf-8')))

survived = np.array([int(row['Survived']) for row in rows])
sex = np.array([row['Sex'] for row in rows])
embarked = np.array([row['Embarked'] if row['Embarked'] else 'Missing' for row in rows])
fare = np.array([float(row['Fare']) for row in rows])
age = np.array([float(row['Age']) if row['Age'] else np.nan for row in rows])

print(len(rows))
print(list(rows[0].keys()))
```

## Part 7: `np.where()` on Titanic Data

Use `np.where()` to label fares as `High Fare` or `Regular Fare`.


```python
fare_labels = np.where(fare > 50, 'High Fare', 'Regular Fare')
print(fare_labels[:5])
```

## Part 8: `np.argmax()` on Titanic Data

Use `np.argmax()` to find the index of the oldest passenger.


```python
oldest_index = np.nanargmax(age)

print(oldest_index)
print(rows[oldest_index]['Name'])
print(age[oldest_index])
```

## Part 9: `np.unique()` on Titanic Data

Check unique values in the `Sex` and `Embarked` columns.


```python
print(np.unique(sex))
print(np.unique(embarked))
```
